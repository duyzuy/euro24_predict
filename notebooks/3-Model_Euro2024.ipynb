```python
import joblib
import pandas as pd
```

```python
# load model 1X2
model_results = \
"../data/euro20_results.joblib"
loaded_model = \
joblib.load(model_results)

# load model goals
model_goals = \
"../data/euro20_goals.joblib"
loaded_model_goals = \
joblib.load(model_goals)

# load data
# matches Group Euro 2024
github = 'https://raw.githubusercontent.com/martj42' 
df_allmatches = pd.read_csv(
f'{github}/international_results/master/results.csv')

# matches Qualifiers Euro 2024
df_qual_24 = pd.read_csv(
    '../data/euro24_qualifiers_norm.csv',
    index_col=0)
```

```python
df_group_24 = df_allmatches[
  (df_allmatches.tournament=='UEFA Euro') & 
  (df_allmatches.date>='2024-06-14') & 
  (df_allmatches.date<='2024-06-26')]
len(df_group_24)
```

```python
group_24_teams = \
list(df_group_24.home_team.unique()) + \
list(df_group_24.away_team.unique())
# eliminate duplicates 
group_24_teams = list(set(group_24_teams))
```

```python
df_qual_24_filter = df_qual_24[
    df_qual_24.team_name.isin(group_24_teams)]

qual_24_teams = \
df_qual_24_filter['team_name'].unique()
len(qual_24_teams)
```

```python
group_24_teams.extend(['Czechia', 'Türkiye'])
df_qual_24_filter = df_qual_24[
    (df_qual_24.team_name.isin(group_24_teams))]
df_qual_24_filter
```

```python
sel_feats = ['goals', 'attempts', 'attempts_on_target', 'attempts_off_target',
             'passes_accuracy', 'passes_attempted',
             'ball_possession', 'cross_accuracy', 'cross_attempted', 
             'free_kick', 'attacks', 'assists', 'corners',
             'offsides', 'recovered_ball', 'tackles', 
             'clearance_attempted', 'saves',
             'goals_conceded', 'fouls_committed', 'fouls_suffered']


df_qual_24_filter[['team_name']+sel_feats]
```

```python
df_group_24.replace('Czech Republic', 'Czechia', inplace=True)
df_group_24.replace('Turkey', 'Türkiye', inplace=True)
df_group_24 = df_group_24[df_group_24['home_team'] != 'Germany']
df_group_24
```

```python
# create dataset 
# home, away, stats_home, stats_away, result (H, D, A)

all_stats = []

# loop on the df_group_24
for idx, row in df_group_24.iterrows():
    
    match_stats = []
    home_team = row['home_team']
    away_team = row['away_team']
    
    home_stats = df_qual_24_filter[df_qual_24_filter['team_name']==home_team].copy()
    away_stats = df_qual_24_filter[df_qual_24_filter['team_name']==away_team].copy()
        
    home_stats = home_stats.add_suffix('_home')
    away_stats = away_stats.add_suffix('_away')
    
    
    match_stats.extend(home_stats.values[0])
    match_stats.extend(away_stats.values[0])
    
    all_stats.append(match_stats)

print(all_stats[0])
```

```python
columns = list(home_stats.columns) + list(away_stats.columns) 
df_matches_stats = pd.DataFrame(all_stats, columns=columns)
df_matches_stats.head()
```

```python
X = df_matches_stats.drop(labels=['team_name_home', 'team_name_away'],
             axis=1)

Y = loaded_model.predict(X)

# decode predictions
predictions = ['X' if pred==0 else str(pred) for pred in Y]

# add back to the matches 
df_matches_stats['result_prediction'] = predictions
# check the slip
df_matches_stats[['team_name_home', 'team_name_away', 'result_prediction']]
```

```python
# extract probabilities
Y_proba = loaded_model.predict_proba(X)
max_prob = [max(prob) for prob in Y_proba]

df_matches_stats['prob'] = max_prob
df_matches_stats[df_matches_stats.prob>=0.6][
    ['team_name_home', 'team_name_away', 
     'result_prediction', 'prob']]
```
