```python
def build_ranking(df):
    
    # build a ranking 
    # rank each column according to its meaning 
    cols = [col for col in df.columns \
            if col not in ['team_name', 'matches_appearance']]
    # print(cols)
    cols_dict = [{'col_name':a, 
                  'ascending':False} for a in cols]
    asc_false = ['offsides', 'goals_conceded', 
                 'own_goal_conceded', 'yellow_cards', 
                 'red_cards', 'fouls_committed']
    
    for d in cols_dict:
        if d['col_name'] in asc_false:
            d['ascending'] = True

    # now rank the dataframe
    df_full_rank = pd.DataFrame()
    for d in cols_dict:
        df_full_rank[d['col_name']] = df[d['col_name']].rank(
            ascending=d['ascending'])

    # make the sum of all the ranks
    df_full_rank['full_rank'] = df_full_rank.sum(axis=1)
    # make the average of all the ranks
    df_full_rank['avg_rank'] = df_full_rank.mean(axis=1)

    # put back the team names
    df_full_rank['team_name'] = df['team_name']
    
    
    
    return df_full_rank
```

```python
df_full_rank = build_ranking(df_norm_qual24_filter)
df_full_rank
```
